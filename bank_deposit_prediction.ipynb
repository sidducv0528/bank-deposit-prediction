```python
out_cols = ['age','campaign','cons.conf.idx']
for col in out_cols:
  q1 = data[col].quantile(0.25)
  q3 = data[col].quantile(0.75)

  iqr = q3-q1
  lb = q1 - (1.5*iqr)
  ub = q3 + (1.5*iqr)
```

**Encoding?**

In machine learning models works only with numeric data. So we need to convert text data to numbers . The process of converting text to numbers is known as encoding

**Label Encoding-** used when we have 2 categories or data is in order

- **how it works?**

-1- find unique values

-2- sort them alphabetically

-3- assign them numbers starting from 0

**Feature selection**

**VIF (Variance Inflation Factor)** - It is a feature selection technique which calculates the multicolinearity between the columns and drops the column with the highest value

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
# create an empty list
col_list = []
for i in data.columns:
  if((data[i].dtype!='object') & (i!='y')):
    col_list.append(i)


X = data[col_list]
vif_data = pd.DataFrame()
vif_data['Column_name'] = X.columns
vif_data['VIF_Score'] = [variance_inflation_factor(X.values,i) for i in range(len(X.columns))]
```

```python
vif_data
```

```python
# VIF SCORE VALUE --- 1 ----(no multicolinarity)
# 1-5 --------------(low multicolinearity)
# 5 or gretear than 5 ----------------(high multicolinearity)
```
